```python
import pandas as pd
import numpy as np
import os
```

```python
# FILL PATHS IN THIS CELL

#Put path to input file
in_path = "E:\TKC_lab_stuff\chromatograpy_plots\Raw_data\Superdex 200\Markers Run Sup 200 BETA AMYL BSA CYTOC 001.csv"
#Put path to folder where output should be saved
path = "E:\TKC_lab_stuff\chromatograpy_plots\Raw_data\Superdex 200" 

in_name = in_path.split("\\")[-1].split(".")[0] 
name = in_name.replace(" ", "_")
out_name = name + '_clean.csv'
out_path = os.path.join(path, out_name)

#IF you want a specific name for your output file, uncomment the next line
#out_path = " insert path to file here "
```

```python
df = pd.read_csv(in_path)
df
```

```python
#creating a dictionary with columns 1 and 2 which contain elution volume and corresponding absorbance values
#change the values in iloc bracket if data looks different 
d ={
    'Volume' : df.iloc[2:,0],
    'Abs' : df.iloc[2:,1]
}
#creating a dataframe from dictionary
col1_2 = pd.DataFrame(data = d, dtype=float)
```

```python
col1_2
```

```python
#finding the index of the first positive elution volume entry
start = col1_2['Volume'].gt(0).idxmax()
start
```

```python
#creating a final dataframe that only has elution volume, absorbance after injection of the sample
final_data = col1_2.iloc[start-2:]
```

```python
final_data
```

```python
final_data.to_csv(out_path, index = False)
```
